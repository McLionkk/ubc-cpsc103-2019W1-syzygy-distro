## Problem 1

Design a function that takes a city name,  minimum grant amount, and a list of gaming grants and returns the gaming grants for the given city of at least the given minimum grant amount.

In particular, the function should operate in two steps. First, it will find the gaming grants for the given city, then it will find which of those gaming grants received at least the minimum grant amount.

Remember to follow the HtDF recipe and the helper rules.

```python



@typecheck
def filter_on_city_and_grant_amount(city: str, amt: int, logg: List[GamingGrant]) -> List[GamingGrant]:
    """
    return a list of grants from logg that are from the given city and of at least amt
    """
    #return [] # body of the stub
    # template based on function composition
    
    return bigger_than_only(amt, from_city_only(city, logg))

@typecheck
def bigger_than_only(amt: int, logg: List[GamingGrant]) -> List[GamingGrant]:
    """
    return a list of grants from logg that are of at least amt
    """
    # return [] # body of the stub
    # template from List[GamingGrant]
    # acc stores the result so far
    acc = [] # type: List[GamingGrant]
    for gg in logg:
        if (grant_bigger_than_x(amt, gg)):
            acc.append(gg)
    return acc

@typecheck
def grant_bigger_than_x(x: int, gg: GamingGrant) -> bool:
    """
    return True if gg has an amount greater than x and False otherwise
    """
    #return False # body of the stub
    # template from GamingGrant 
    return gg.amt > x

@typecheck
def from_city_only(city: str, logg: List[GamingGrant]) -> List[GamingGrant]:
    """
    return a list of grants from logg that are from city
    """
    # return [] # body of the stub
    # template from List[GamingGrant]
    # acc stores the result so far
    acc = [] # type: List[GamingGrant]
    for gg in logg:
        if (from_city(city, gg)):
            acc.append(gg)
    return acc

@typecheck
def from_city(city: str, gg: GamingGrant) -> bool:
    """
    return True if gg is from the given city, False otherwise
    """
    #return False # body of the stub
    # template from GamingGrant 
    return gg.city == city

start_testing()

expect(filter_on_city_and_grant_amount("Burnaby", 20000, []), []) 
expect(filter_on_city_and_grant_amount("Burnaby", 20000, L1), [BARRACUDAS])
expect(filter_on_city_and_grant_amount("Burnaby", 2000, L1), [ARTS_COUNCIL, BARRACUDAS])
expect(filter_on_city_and_grant_amount("Burnaby", 15099, L1), [ARTS_COUNCIL, BARRACUDAS])
expect(filter_on_city_and_grant_amount("Richmond", 30000, L1), [YOUTH_MUSIC])
expect(filter_on_city_and_grant_amount("Richmond", 50000, L1), [])

expect(bigger_than_only(1, []), [])
expect(bigger_than_only(4999, L1), L1)
expect(bigger_than_only(30000, L1), [AGRIFAIR, AIRSHOW, BARRACUDAS, YOUTH_MUSIC])
expect(bigger_than_only(85000, L1), [])

expect(grant_bigger_than_x(3000, CHINESE_ORCHESTRA), True)
expect(grant_bigger_than_x(5000, CHINESE_ORCHESTRA), False)
expect(grant_bigger_than_x(10000, CHINESE_ORCHESTRA), False)
expect(grant_bigger_than_x(29499, BMX), True)
expect(grant_bigger_than_x(29500, BMX), False)
expect(grant_bigger_than_x(29501, BMX), False)

expect(from_city_only("Burnaby", []), []) 
expect(from_city_only("Burnaby", L1), [ARTS_COUNCIL, BARRACUDAS])
expect(from_city_only("Coquitlam", L1), [])
expect(from_city_only("Richmond", L1), [YOUTH_MUSIC, CHINESE_ORCHESTRA])

expect(from_city("Richmond", CHINESE_ORCHESTRA), True)
expect(from_city("Vancouver", CHINESE_ORCHESTRA), False)
expect(from_city("Abbotsford", BMX), True)
expect(from_city("North Vancouver", BMX), False)

summary()
```
